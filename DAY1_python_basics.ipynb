```python
tom_exp_list = [2100, 3400, 3500]
joe_exp_list = [200, 500, 700]

total = 0
for item in tom_exp_list:
  total = total + item
print("Tom's total expense: ", total)

total = 0
for item in joe_exp_list:
  total = total + item
print("Joe's total expense: ", total)
```

```python
def calculate_total(exp):
  total = 0
  for item in exp:
    total = total + item
  return total
```

```python
toms_total = calculate_total(tom_exp_list)
joes_total = calculate_total(joe_exp_list)
print("Toms total exp: ", toms_total)
print("joes total exp: ", joes_total)
```
