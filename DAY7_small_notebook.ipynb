```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
df['species'] = iris.target
```

```python
df
```

```python
arr = np.array(iris.data)
print(arr.mean(axis=0))
print(arr.max(axis=0))
print(arr.shape)
print(arr[0:5])
```

```python
df.head()
df.info()
df.describe()
df[df['species']==0]
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
df.groupby('species')['petal length (cm)'].mean().plot(kind = 'bar')
plt.title('avg petal length by species')
plt.show()
```

```python
plt.scatter(df['petal length (cm)'], df['petal width (cm)'], c=df['species'])
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.title('Sepal vs petal lenggth')
plt.show()
```
