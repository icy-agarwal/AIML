```python
import time
```

```python
SIZE = 1000

L1 = range(SIZE)
L2 = range(SIZE)

a1 = np.arange(SIZE)
a2 = np.arange(SIZE)

#python list
start = time.time()
result = [(x+y) for x, y in zip(L1, L2)]
print("python list took: ", (time.time()-start)*1000)

#python array
start = time.time()
result = a1 + a2
print("numpy took: ", (time.time()-start)*1000)
```
