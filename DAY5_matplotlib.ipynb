```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.xlabel('day')
plt.ylabel('temp')
plt.title('weather')
plt.plot(x,y, color = 'green', linewidth=5, linestyle = 'dashdot')
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.plot(days, max_t, label = "max")
plt.plot(days, min_t, label = "min")
plt.plot(days, avg_t, label = "avg")
plt.xlabel("days")
plt.ylabel("temp")
plt.title("weather")
plt.legend(loc="best")
plt.grid()
```

```python
company = ['google', 'amazon', 'msft', 'facebook']
revenue = [90, 136, 59, 27]
```

```python
xpos = np.arange(len(company))
```

```python
ypos = np.arange(len(company))
```

```python
xpos
```

```python
plt.xticks(xpos, company)
plt.bar(xpos-0.2, revenue, label = "revenue", width = 0.4)
profit = [40, 2, 34, 12]
plt.bar(xpos+0.2, profit, label = "profit", width = 0.4)
plt.legend()
```

```python
plt.yticks(ypos, company)
plt.title("us tech stocks")
plt.barh(ypos-0.2, revenue, label = "revenue")
plt.barh(ypos+0.2, profit, label = "profit")
plt.legend()
```
